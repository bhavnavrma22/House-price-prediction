```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
def load_data(path):
    """Loads CSV data"""
    return pd.read_csv(path)

def train_model(X, y):
    """Trains a Linear Regression model"""
    model = LinearRegression()
    model.fit(X, y)
    return model

def predict_price(model, area):
    """Predicts price for a given area"""
    return model.predict([[area]])

def evaluate_model(model, X, y):
    """Prints MSE and R2 Score"""
    predictions = model.predict(X)
    mse = mean_squared_error(y, predictions)
    r2 = r2_score(y, predictions)
    print(f"Mean Squared Error: {mse:.2f}")
    print(f"R² Score: {r2:.2f}")
```

```python
# Load the data
data = load_data('dataset.csv')
print(data.head())
```

```python
# Prepare training data
X = data[['Area']]
y = data['Price']

# Train the model
model = train_model(X, y)
```

```python
# Evaluate the model
evaluate_model(model, X, y)
```

```python
# Visualize the regression line
plt.scatter(data['Area'], data['Price'], color='blue', label='Actual')
plt.plot(data['Area'], model.predict(X), color='red', label='Predicted')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price')
plt.title('House Price Prediction')
plt.legend()
plt.show()
```
